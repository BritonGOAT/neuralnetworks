```python
# Imports math library
import numpy as np
# Imports plotting library
import matplotlib.pyplot as plt
```

```python
# Define a linear function with just one input, x
def linear_function_1D(x,beta,omega):
  y = beta + omega * x

  return y
```

```python
 #Plot the 1D linear function

# Define an array of x values from 0 to 10 with increments of 0.01
# https://numpy.org/doc/stable/reference/generated/numpy.arange.html
x = np.arange(0.0,10.0, 0.01)
# Compute y using the function you filled in above
beta = 10.0; omega = 2.0

y = linear_function_1D(x,beta,omega)

# Plot this function
fig, ax = plt.subplots()
ax.plot(x,y,'r-')
ax.set_ylim([0,20]);ax.set_xlim([0,20])
ax.set_xlabel('x'); ax.set_ylabel('y')
plt.show

# TODO -- experiment with changing the values of beta and omega
# to understand what they do.  Try to make a line
# that crosses the y-axis at y=10 and the x-axis at x=5
```

```python
# Code to draw 2D function -- read it so you know what is going on, but you don't have to change it
def draw_2D_function(x1_mesh, x2_mesh, y):
    fig, ax = plt.subplots()
    fig.set_size_inches(7,7)
    pos = ax.contourf(x1_mesh, x2_mesh, y, levels=256 ,cmap = 'hot', vmin=-10,vmax=10.0)
    fig.colorbar(pos, ax=ax)
    ax.set_xlabel('x1');ax.set_ylabel('x2')
    levels = np.arange(-10,10,1.0)
    ax.contour(x1_mesh, x2_mesh, y, levels, cmap='winter')
    plt.show
```

```python
# Define a linear function with two inputs, x1 and x2
def linear_function_2D(x1,x2,beta,omega1,omega2):
  # TODO -- replace the code line below with formula for 2D linear equation
   y = beta + omega1 * x1 + omega2 * x2
   return y
```

```python
# Define a linear function with three inputs, x1, x2, and x_3
def linear_function_3D(x1,x2,x3,beta,omega1,omega2,omega3):
  # TODO -- replace the code below with formula for a single 3D linear equation
   y = beta + omega1 * x1 + omega2 * x2 + omega3 * x3
   return y
```

```python
import numpy as np

# Define the parameters
beta1 = 0.5; beta2 = 0.2
omega11 =  -1.0 ; omega12 = 0.4; omega13 = -0.3
omega21 =  0.1  ; omega22 = 0.1; omega23 = 1.2

# Define the inputs
x1 = 4 ; x2 =-1; x3 = 2

# Define linear_function_3D
def linear_function_3D(x1, x2, x3, beta, omega1, omega2, omega3):
    return beta + omega1 * x1 + omega2 * x2 + omega3 * x3

# Compute using the individual equations
y1 = linear_function_3D(x1, x2, x3, beta1, omega11, omega12, omega13)
y2 = linear_function_3D(x1, x2, x3, beta2, omega21, omega22, omega23)

print("Individual equations")
print('y1 = %3.3f\ny2 = %3.3f' % (y1, y2))

# Define vectors and matrices
beta_vec = np.array([[beta1], [beta2]])
omega_mat = np.array([[omega11, omega12, omega13], [omega21, omega22, omega23]])
x_vec = np.array([[x1], [x2], [x3]])

# Compute with vector/matrix form
y_vec = beta_vec + np.matmul(omega_mat, x_vec)

print("Matrix/vector form")
print('y1 = %3.3f\ny2 = %3.3f' % (y_vec[0, 0], y_vec[1, 0]))
```

```python
#A single linear equation with three inputs (i.e. linear_function_3D()) associates a
#value y with each point in a 3D space (0,0,0,). 
#Is it possible to visualize this? What value is at position (0,0,0)?

#value at (0,0,0) is y=beta value()

from mpl_toolkits.mplot3d import Axes3D

# Define the inputs
x1 = np.linspace(-5, 5, 100)
x2 = np.linspace(-5, 5, 100)
x1_mesh, x2_mesh = np.meshgrid(x1, x2)
x3 = 2  # Constant value for this visualization

# Define linear_function_3D
def linear_function_3D(x1, x2, x3, beta, omega1, omega2, omega3):
    return beta + omega1 * x1 + omega2 * x2 + omega3 * x3

# Compute the values
y1 = linear_function_3D(x1_mesh, x2_mesh, x3, beta1, omega11, omega12, omega13)
y2 = linear_function_3D(x1_mesh, x2_mesh, x3, beta2, omega21, omega22, omega23)

# Create the 3D plot
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
surf1 = ax.plot_surface(x1_mesh, x2_mesh, y1, cmap='coolwarm', alpha=0.7)
surf2 = ax.plot_surface(x1_mesh, x2_mesh, y2, cmap='viridis', alpha=0.7)

# Add labels and title
ax.set_xlabel('X1')
ax.set_ylabel('X2')
ax.set_zlabel('Y')
ax.set_title('Surface plot of linear_function_3D')

# Add a color bar which maps values to colors
fig.colorbar(surf1, ax=ax, shrink=0.5, aspect=5)

# Show the plot
plt.show()
```

```python
#Write code to compute three linear equations with two inputs (x1,x2) using both the individual equations 
#and the matrix form (you can make up any values for the inputs bi and the slopes wij.


# Define inputs and slopes
x1 = np.array([2, 3, 4])  #  x1
x2 = np.array([5, 6, 7])  #  x2

# Individual equations
def linear_equation_1(x1, x2):
    return 3 * x1 + 4 * x2

def linear_equation_2(x1, x2):
    return -2 * x1 + 5 * x2

def linear_equation_3(x1, x2):
    return 4 * x1 - 3 * x2

# Compute values using individual equations
y1 = linear_equation_1(x1, x2)
y2 = linear_equation_2(x1, x2)
y3 = linear_equation_3(x1, x2)

print("Values using individual equations:")
print("y1:", y1)
print("y2:", y2)
print("y3:", y3)

# Matrix form
coeff_matrix = np.array([[3, 4], [-2, 5], [4, -3]])  # Coefficient matrix
input_matrix = np.column_stack((x1, x2))  # Input matrix

# Compute values using matrix form
y_matrix = np.dot(input_matrix, coeff_matrix.T)

print("\nValues using matrix form:")
print("y_matrix:", y_matrix)
```
